```python
import random
from pathlib import Path

def get_random_files(base_path, pattern="*.txt", num_samples=1):
    files = list(Path(base_path).glob(pattern))  # Buscar archivos con el patr√≥n dado
    return random.sample(files, min(num_samples, len(files)))  # Seleccionar aleatorios

# Ejemplo de uso
random_files = get_random_files("C:/Users/opera/OneDrive/Documentos/GitHub/BreastCATT/BreastThermography", "**/*.txt", 1)
print(random_files)
```

```python
import numpy as np
import matplotlib.pyplot as plt

def plot_thermal_image(txt_path):
    thermal_matrix = np.loadtxt(txt_path)  # Cargar la matriz de temperatura
    plt.imshow(thermal_matrix, cmap='hot', interpolation='nearest')  # Usar colormap 'hot'
    plt.colorbar(label='Temperature (¬∞C)')  # Agregar barra de colores
    plt.title('Thermal Image:')
    plt.show()
    return thermal_matrix

# Uso: Especifica la ruta de un archivo .txt generado

thermal_matrix = plot_thermal_image(str(random_files[0]))  
print(f"max temp: {np.max(thermal_matrix)}")
print(f"min temp: {np.min(thermal_matrix)}")
```
