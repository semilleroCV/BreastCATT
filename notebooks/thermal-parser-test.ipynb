"thermal parser" es un repositorio que ayuda en la transformacion de imagenes JPEG de tipo "lossless radiometric JPEG", haciendo el proceso de calibracion entre nuestra imagen comprimida a un txt de temperatura, esto es posible gracias a una herramienta llamada [exiftool](https://exiftool.org/)

```python
from thermal_parser import Thermal
from pathlib import Path
import numpy as np
```

```python
path = Path(r"C:\Users\opera\OneDrive\Documentos\GitHub\BreastCATT\BreastThermography\test\malignant\IIR0043_anterior.jpg")

thermal = Thermal(dtype=np.float32)
temperature = thermal.parse(filepath_image=str(path))
assert isinstance(temperature, np.ndarray)
```

```python
import matplotlib.pyplot as plt
# dimensiones de la imagen
print(f"Dimensiones de la imagen: {temperature.shape}")
print(f"Valor mínimo: {temperature.min()}")
print(f"Valor máximo: {temperature.max()}")

plt.figure(figsize=(12, 8))
plt.imshow(temperature, cmap="hot")
plt.axis("off")
plt.colorbar()
plt.title("Imagen térmica TIFF")
plt.show()
```
